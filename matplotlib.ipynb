```python
%pip install --quiet matplotlib

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(-10, 10, 21)
y = x**2
plt.plot(x, y,'g-')
plt.title('y = x^2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
# green line plot of y = x^2
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'g-')
plt.subplot(1, 2, 2)
plt.plot(x, y, 'r-')
# red line plot of y = x^2 & green line plot of y = x^2
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2)
axes[0, 0].plot(x, y, 'g-')
axes[0, 1].plot(y, y, 'r-')
axes[1, 0].plot(y, x, 'b-')
axes[1, 1].plot(x**-1, y, 'y-')
# 2x2 grid of subplots with different line plots
```

```python
fig.savefig('line_plot.png')
# save the figure as a PNG file
```

```python
fig = plt.figure(figsize=(10, 5))
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, label='y = x^2')
axes.plot(y, x, label='y = x^2')
axes.legend()
```
